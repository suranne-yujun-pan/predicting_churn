```python
columns_to_drop2 = ['state', 'area_code']
```

```python
train = train.drop(columns_to_drop2, axis=1)
```

```python
#telco['Intl_Plan'] = telco['Intl_Plan'].replace({'no': 0, 'yes':1})
#train['international_plan'] = train['international_plan'].replace({'no':0, 'yes':1}) 
```

```python
train['voice_mail_plan'] = train['voice_mail_plan'].replace({'no':0, 'yes':1}) 
```

```python
# Import necessary modules
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
X = train.drop('churn', axis=1)
```

```python
y = train['churn']
```

```python
# Create training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 42)
```

```python
# Instantiate the classifier
clf = RandomForestClassifier()
```

```python
# Fit to the data
clf.fit(X_train, y_train)
```

```python
# Print the accuracy
print(clf.score(X_test, y_test))
```

```python
from sklearn.metrics import confusion_matrix
y_pred = clf.predict(X_test)
#print(confusion_matrix(y_test, y_pred, labels=[0,1]))
```

```python
print(confusion_matrix(y_test, y_pred, labels=['no','yes']))
```

```python
print('sensitivity :', 138/(57+138))
```

```python
# Import recall_score
from sklearn.metrics import recall_score

# Print the recall
print(recall_score(y_test, y_pred, average="binary", pos_label="yes"))
```

```python
# Import f1_score
from sklearn.metrics import f1_score

# Print the F1 score
print(f1_score(y_test, y_pred, average="binary", pos_label="yes"))
```
